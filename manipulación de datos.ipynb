# Paso 3: Limpieza de datos

### En este paso revisamos qué hacer con los datos que "ensucian" el dataset.



```python
# Obtenemos el porcentaje de datos nulos 
df_mental_health.isnull().sum() / df_mental_health.shape[0] * 100
```

```python
# Obtenemos el número total de datos nulos
total_nulos = df_mental_health.isnull().sum().sum()

# Obtenemos el número total de datos en el dataframe
total_datos = df_mental_health.size

# Obtenemos el porcentaje de datos nulos
print(f"{round(total_nulos / total_datos * 100, 2)}% de los datos son nulos" )
```

Dado que las columnas Medication_Use y Substance_Use tienen valores nulos o faltantes, se opta por eliminar estas columnas del dataset al ser las únicas que contienen datos inconsistentes. Si bien podríamos optar por rellenar los valores nulos con la moda, en este caso no es necesario ya que el dataset tiene un tamaño considerable y no se perdería información relevante al eliminar estas columnas o el contexto del dataset puede desestimar la información de estas columnas.

```python
df_mental_health_clean = df_mental_health.drop(['Medication_Use','Substance_Use'], axis=1)
df_mental_health_clean.head()
```

# Paso 4: Análisis estadístico
En este paso, se realizan estadísticas básicas para las columnas Anxiety_Score, Depression_Score y Stress_Level.

```python
# Anxiety_Score
Anxiety_Score_mean = df_mental_health_clean['Anxiety_Score'].mean()
Anxiety_Score_median = df_mental_health_clean['Anxiety_Score'].median()
Anxiety_Score_std = df_mental_health_clean['Anxiety_Score'].std()

print("Anxiety_Score")
print("Mean:", Anxiety_Score_mean)
print("Median:", Anxiety_Score_median)
print("Std:", Anxiety_Score_std)
```

```python
# Depression_Score
Depression_Score_mean = df_mental_health_clean['Depression_Score'].mean()
Depression_Score_median = df_mental_health_clean['Depression_Score'].median()
Depression_Score_std = df_mental_health_clean['Depression_Score'].std()

print("Depression_Score")
print("Mean:", Depression_Score_mean)
print("Median:", Depression_Score_median)
print("Std:", Depression_Score_std)
```

```python
# Stress_Level
Stress_Level_mean = df_mental_health_clean['Stress_Level'].mean()
Stress_Level_median = df_mental_health_clean['Stress_Level'].median()
Stress_Level_std = df_mental_health_clean['Stress_Level'].std()

print("Stress_Level")
print("Mean:", Stress_Level_mean)
print("Median:", Stress_Level_median)
print("Std:", Stress_Level_std)
```

Con esta información se puede observar que:

- La variable Anxiety_Score tiene la desviación estándar más alta, lo que indica que tiene una mayor variabilidad en comparación con las otras dos variables con 5.91 
- Dada la similitud entre la media y la mediana de las tres variables esto sugiere que la distribución de los datos es relativamente simétrica.

# Paso 5: Filtrado de datos

```python
puntaje_alto_ansiedad = df_mental_health_clean['Anxiety_Score'].max() * 0.8
personas_ansiedad_alto = df_mental_health_clean[df_mental_health_clean['Anxiety_Score'] > puntaje_alto_ansiedad]

personas_ansiedad_alto.head(10)
```

```python
# Se calcula el porcentaje de personas con ansiedad alta
porcentaje_ansiedad_alto = (len(personas_ansiedad_alto) / len(df_mental_health_clean)) * 100
print(f"El {porcentaje_ansiedad_alto:.2f}% de las personas tienen ansiedad alta.")
```

```python
# Buscamos si existe correlación entre el la variable Stress_Level y Sleep_Hours
correlacion = df_mental_health_clean['Stress_Level'].corr(df_mental_health_clean['Sleep_Hours'])
print(f"La correlación entre Stress_Level y Sleep_Hours es: {correlacion:.2f}")
```

# Paso 6: Ordenamiento



```python
df_mental_health_sorted = df_mental_health_clean.sort_values(by=['Life_Satisfaction_Score', 'Loneliness_Score'], ascending=[False,True])
df_mental_health_sorted.head(10)
```

```python
# Mostramos solos las columnas de interés Life_Satisfaction_Score y Loneliness_Score
df_mental_health_sorted[['Life_Satisfaction_Score', 'Loneliness_Score']].head(10)

# Buscamos la correlación entre Life_Satisfaction_Score y Loneliness_Score
correlacion = df_mental_health_sorted['Life_Satisfaction_Score'].corr(df_mental_health_sorted['Loneliness_Score'])
print(f"La correlación entre Life_Satisfaction_Score y Loneliness_Score es: {correlacion:.2f}")
```

Con esta información se puede observar que:
- Al analizar las primera 10 personas con los valores ordenados, pareciera surgir un patrón en el que las personas con mayor puntaje de satisfacción de vida tiene un menor puntaje de soledad.
- Si se busca la correlación se descubre que esto no es así


# Paso 7: Guardado de Resultados

```python
# Se filtra el dataset para encontrar a las personas que reciben terapia y que además tienen un nivel bajo de estrés (por debajo de la mediana)

# Obtenemos la mediana de Stress_Level
mediana_ansiedad = df_mental_health_clean['Stress_Level'].median()

# Se crea un nuevo dataframe con las personas que reciben terapia y que tienen un nivel ed estrés por debajo de la mediana 
df_terapia_bajo_estres = df_mental_health_clean[(df_mental_health_clean['Therapy'] == 1) & (df_mental_health_clean['Stress_Level'] < mediana_ansiedad)]

df_terapia_bajo_estres.shape

#df_terapia_bajo_estres.head(10)
```

```python
# Buscamos la correlación entre Therapy y Stress_Level
correlacion = df_mental_health_clean['Therapy'].corr(df_mental_health_clean['Stress_Level'])
print(f"La correlación entre Therapy y Stress_Level es: {correlacion:.2f}")
```

Con esta información se puede observar que:
- Hay 108 personas que cumple con la condición de recibir terapia y además tienen un nivel de estrés por debajo de la mediana.

# Paso 8: Visualización Opcional



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Se crea un gráfico para mostrar la media de la variable Self_Esteem_Score agrupado por la variable Gender

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_mental_health_clean,
    x='Gender',
    y='Self_Esteem_Score',
    ci=None,
    palette='viridis'
)
plt.title('Media de Self_Esteem_Score por Gender')
plt.xlabel('Gender')
plt.ylabel('Self_Esteem_Score (Media)')
plt.show()
```

Apoyándonos en la visualización de los datos, se puede observar que:

- No hay diferencias notables entre los géneros en cuanto a la autoestima
